#### Arithmetic Operators

| Operator | Description         | Example        | Result |
|----------|---------------------|----------------|--------|
| +        | Addition            | SELECT 10 + 5; | 15     |
| -        | Subtraction         | SELECT 10 - 5; | 5      |
| *        | Multiplication      | SELECT 10 * 5; | 50     |
| /        | Division            | SELECT 10 / 5; | 2      |
| %        | Modulus (remainder) | SELECT 10 % 3; | 1      |

1. Calculate the Annual Salary and Salary Increment by 5% - show the monthly new salary as well


```sql
SELECT first_name, last_name, salary, 
    (salary * 12) AS annual_salary,
    (salary * .05) AS increment_amount,
    (salary * 1.05 AS new_salary,
    (salary + salary * .05) AS new_salary2
FROM ex_employees;    
```

````python
%%sql

SELECT first_name, last_name, salary,
        (salary * 12) AS annual_salary,
        (salary * .05) AS increment_amount,
        (salary + salary * .05) AS new_salary,
        (salary * 1.05) AS new_salary2
FROM ex_employee;
````

#### Comparison Operators

| Operator | Description           |
|----------|-----------------------|
| =        | Equal to              |
| != or <> | Not equal to          |
| >        | Greater than          |
| <        | Less than             |
| >=       | Greater than or equal |
| <=       | Less than or equal    |

````python
%%sql
SELECT * FROM ex_employee
WHERE age=30;
````

````python
%%sql
SELECT * FROM ex_employee
WHERE age!=30;

-- SELECT * FROM ex_employee
-- WHERE age<>30;
````

````python
%%sql
SELECT * FROM ex_employee
WHERE salary > 50000;
````

````python
%%sql
SELECT * FROM ex_employee
WHERE salary >= 50000;
````

````python
%%sql
SELECT * FROM ex_employee
WHERE salary < 50000;
````

````python
%%sql
SELECT * FROM ex_employee
WHERE salary <= 50000;
````

#### Logical Operators

| Operator | Description                         |
|----------|-------------------------------------|
| AND      | All conditions must be true         |
| OR       | At least one condition must be true |
| NOT      | Negates the condition               |

````python
%%sql
SELECT * FROM ex_employee
WHERE age >= 40 AND salary >= 5000;
````

````python
%%sql
SELECT * FROM ex_employee
WHERE age > 60 or salary >= 90000;
````

````python
%%sql
SELECT * FROM ex_employee
WHERE NOT (department = 'IT');
````

#### Between, Like and In Operators

1. Retrieve employees whose salary is between 40,000 and 60,000. - Use BETWEEN Operators

````python
%%sql
SELECT first_name, last_name, salary FROM ex_employee
WHERE salary BETWEEN 40000 AND 60000;

-- SELECT first_name, last_name, salary FROM ex_employee
-- WHERE salary >= 40000 AND salary <= 60000;
````

2. Find employee whose email addresses end with gmail.com - Use LIKE Operators

````python
%%sql

SELECT first_name, last_name, salary FROM ex_employee
WHERE email LIKE '%gmail.com';
````

````python
%%sql
SELECT first_name, salary FROM ex_employee
WHERE first_name LIKE 'J%' ORDER BY salary ASC;
````

3. Retrieve employees who belong to either the 'Finance' or 'Marketing' departments -- Use IN Operator

````python
%%sql
SELECT first_name, last_name, department FROM ex_employee
WHERE department IN ('Finance', 'Marketing');

-- SELECT first_name, last_name, department FROM ex_employee
-- WHERE department = 'Finance' OR department = 'Marketing';
````

#### Other Operators

1. Find employees where the email column is NULL (if applicable)

````python
%%sql
SELECT first_name, last_name, email FROM ex_employee
WHERE email IS NULL;
````

2. List employees sorted by salary in DESCENDING order.

````python
%%sql
SELECT first_name, last_name, salary FROM ex_employee
ORDER BY salary DESC;

-- SELECT first_name, last_name, salary FROM ex_employee
-- ORDER BY salary ASC;
-- OR
-- ORDER BY salary;
````

3. Retrieve the top 5 highest-paid employees.

````python
%%sql
SELECT first_name, last_name, salary FROM ex_employee
LIMIT 4;
````

````python
%%sql
SELECT first_name, last_name, salary FROM ex_employee
ORDER BY salary DESC
LIMIT 5;
````

4. Retrieve a list of unique departments

````python
%%sql
SELECT DISTINCT department
FROM ex_employee;
````

````python
%%sql
SELECT COUNT(DISTINCT department) as DEPARTMENT_UNIQUE_COUNT
FROM ex_employee;
````

#### Set Operators

| Operator  | Description                          |
|-----------|--------------------------------------|
| UNION     | Combines results, removes duplicates |
| UNION ALL | Combines results, keeps duplicates   |
| INTERSECT | Returns common results               |
| EXCEPT    | Returns results in first, not second |

````python
%%sql
DROP TABLE IF EXISTS students_2026;
````

````python
%%sql
CREATE TABLE students_2026 (
    student_id INT PRIMARY KEY,
    student_name VARCHAR(100),
    course VARCHAR(50)
);
````

````python
%%sql
INSERT INTO students_2026 (student_id, student_name, course) VALUES
(1, 'Aarav Sharma', 'Computer Science'),
(2, 'Ishita Verma', 'Machanical Engineering'),
(3, 'Kabir Patel', 'Electronics'),
(4, 'Ananya Desai', 'Civil Engineering'),
(5, 'Rahul Gupta', 'Computer Science');
````

````python
%%sql
SELECT * FROM students_2026;
````

````python
%%sql
DROP TABLE IF EXISTS students_2025;
````

````python
%%sql
CREATE TABLE students_2025 (
    student_id INT PRIMARY KEY,
    student_name VARCHAR(100),
    course VARCHAR(50)
);
````

````python
%%sql
INSERT INTO students_2025 (student_id, student_name, course) VALUES
(3, 'Kabir Patel', 'Electronics'), -- same as students_2026
(4, 'Ananya Desai', 'Civil Engineering'), -- same as students_2026
(6, 'Meera Rao', 'Computer Science'),
(7, 'Vikram Singh', 'Mathematics'),
(8, 'Sanaya Kapoor', 'Physics');
````

````python
%%sql
SELECT * FROM students_2025;
````

 **-- UNION --** Combines results, removes duplicates

````python
%%sql
SELECT * FROM students_2025
UNION
SELECT * FROM students_2026;
````

**-- UNION ALL --** - Combines results, keeps duplicates

````python
%%sql
SELECT * FROM students_2025
UNION ALL
SELECT * FROM students_2026;
````

**-- INTERSECT --** - Returns common results in both tables

````python
%%sql
SELECT * FROM students_2025
INTERSECT
SELECT * FROM students_2026;
````

**-- EXCEPT --** - Returns results in the first table but not in the second

````python
%%sql
SELECT * FROM students_2025
EXCEPT
SELECT * FROM students_2026;
````
