```python
%%sql
CREATE TABLE IF NOT EXISTS student (
    id SERIAL primary key,
    name VARCHAR(50) NOT NULL,
    email VARCHAR(100) NOT NULL UNIQUE,
    age INT,
    city VARCHAR(50)
);
```

```python
%%sql
SELECT * FROM student;
```
