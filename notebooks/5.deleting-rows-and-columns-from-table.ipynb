```python
%%sql
CREATE TABLE student (
    id INT primary key,
    name VARCHAR(100) not null,
    class VARCHAR(20),
    roll NUMERIC(10)
);
```

```python
%%sql
SELECT * FROM student;
```

```python
%%sql
INSERT INTO student (id, name, class, roll)
VALUES (101, 'Gaurav', 'Post Graduation', 1), 
    (1022, 'Saurav', 'MBBS', 2), 
    (1023, 'Sarita', 'CA', 3);

-- id can't be duplicate
```
