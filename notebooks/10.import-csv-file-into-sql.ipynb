```python
%%sql

DROP TABLE IF EXISTS ex_employee;

CREATE TABLE IF NOT EXISTS ex_employee (
    employee_id	INT primary key,
    first_name VARCHAR(20) NOT NULL,
    last_name VARCHAR(20) NOT NULL,
    email VARCHAR(100) UNIQUE,
    department VARCHAR(20),
    salary NUMERIC(10, 2),
    joining_date TEXT,
    age INT
);

SELECT * FROM ex_employee;
```

```python
%%sql
-- use `/copy` if using jupyter notebook, but since we are using docker, we have to past the .csv file inside postgreSQL itself
copy ex_employee(
  employee_id,
  first_name,
  last_name,
  email,
  department,
  salary,
  joining_date,
  age
)
FROM '/csv/employee_data.csv'
DELIMITER ','
CSV HEADER;
```

```python
%%sql
UPDATE ex_employee
SET joining_date = TO_CHAR(
    TO_DATE(joining_date, 'DD-MM-YYYY'),
    'YYYY-MM-DD'
);
```
