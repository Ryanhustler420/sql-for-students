## Solutions

1. Retrieve all humans first_name and their department.

```python
%%sql
select first_name, department from humans;
```

2. Update the salary of all humans in the `IT` department by increasing it by 10%

```python
%%sql
UPDATE humans
SET salary=salary+salary*.10
WHERE department='IT'
```

```python
%%sql
SELECT * FROM humans;
```

3. Delete all humans who are older than 34 years.

```python
%%sql
DELETE FROM humans
WHERE age > 34;
```

```python
%%sql
SELECT * FROM humans;
```

4. Add a new column `email` to the `humans` table.

```python
%%sql
ALTER TABLE humans
ADD COLUMN email VARCHAR(50);
```

```python
%%sql
SELECT * FROM humans;
```

5. Rename the department column to dept_name.

```python
%%sql
ALTER TABLE humans
RENAME COLUMN department TO dept_name;
```

```python
%%sql
SELECT * FROM humans;
```

6. Retrieve the names of humans who joined after January 1, 2021.

```python
%%sql
SELECT first_name, last_name, joining_date FROM humans
WHERE joining_date > '2021-01-01 00:00:00';
-- WHERE joining_date > '2021-01-01 00:00:00'; -- works too
```

7. Change the data type of the `salary` column to `INTEGER`

```python
%%sql
ALTER TABLE humans
ALTER COLUMN salary TYPE INTEGER;
-- ALTER COLUM salary TYPE INTEGER USING salary::INTEGER
```

```python
%%sql
SELECT * FROM humans;
```

8. List all humans with their age and salary in descending order of salary.

```python
%%sql
SELECT first_name, age, salary FROM humans
ORDER BY salary DESC;
```
